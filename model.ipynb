```python
%pip install ollama

import subprocess
import time
import requests
import os
import ollama

# 1. Install Ollama Engine
def install_ollama():
    print("Checking for Ollama...")
    if subprocess.run(["which", "ollama"], capture_output=True).returncode != 0:
        print("Installing Ollama engine on Colab...")
        !apt-get update && apt-get install -y zstd
        !curl -fsSL https://ollama.com/install.sh | sh
    else:
        print("Ollama engine already installed.")

# 2. Start Server in Background
# We use 'nohup' so the server doesn't die when the cell finishes
def start_ollama_server():
    print("Starting Ollama server...")
    with open("ollama.log", "w") as log:
        subprocess.Popen(["ollama", "serve"], stdout=log, stderr=log, preexec_fn=os.setsid)
    # 3. Wait for Server to be 'Healthy'
    for _ in range(10):
        try:
            if requests.get("http://localhost:11434/api/tags").status_code == 200:
                print("✅ Ollama Server is UP!")

                break
        except:
            time.sleep(2)
    else:
        print("❌ Server timed out. Check 'ollama.log'")

# 4. Install the Model (Llama 3 8B)
def install_model(modle_name="llama3:8b"):
    print(f"Downloading {modle_name} (this may take 2-3 minutes)...",)
    !ollama pull modle_name
    print("✅ Model installed and ready for queries!")

def run_research_query(prompt):
    models_response = ollama.list()
    
    # Access as attributes (.model), not keys (['name'])
    existing_models = [m.model for m in models_response.models]
    print(f"Available models: {existing_models}")
    
    response = ollama.chat(model='llama3:8b', messages=[
        {'role': 'user', 'content': prompt}
    ])
    
    # Also access the response content as an attribute
    return response.message.content

install_ollama()
start_ollama_server()
install_model()
abstract = "get x,y cordinates of an elements in home page of amazon website"
summary = run_research_query(f"{abstract}")
print(summary)
```
